```python
import json
from pathlib import Path

regions_root = Path('regions')

structure = {}
for country_dir in sorted(regions_root.iterdir()):
    if not country_dir.is_dir():
        continue
    regions = sorted(p.name for p in country_dir.iterdir() if p.is_dir())
    if regions:
        structure[country_dir.name] = regions

output_path = Path('region_structure.json')
output_path.write_text(json.dumps(structure, indent=2))

print(f"Written to {output_path}")
for country, regs in structure.items():
    print(f"  {country}: {len(regs)} region(s)")
```
