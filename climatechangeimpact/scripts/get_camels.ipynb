```python
import os
import json
import pandas as pd
```

```python
# Path to file containing camel_ids (one per line)
camel_ids_file = "camel_ids.txt"

# Base directory
base_dir = "../regions"

# Make sure the base directory exists
os.makedirs(base_dir, exist_ok=True)

empty_settings = dict()

# Read camel_ids and create directories
with open(camel_ids_file, "r") as f:
    for line in f:
        camel_id = line.strip()  # remove newline and whitespace
        if camel_id:  # skip empty lines
            dir_path = os.path.join(base_dir, camel_id)
            os.makedirs(dir_path, exist_ok=True)
            print(f"Created: {dir_path}")
            with open(f"{dir_path}/settings.json", "w") as json_file:
                json.dump(empty_settings, json_file, indent=4)
            print(f"created settingsfile: {dir_path}/settings.json")
```
